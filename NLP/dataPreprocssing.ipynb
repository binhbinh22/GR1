```python
import csv
from underthesea import ner
from underthesea import word_tokenize

```

```python

input_file = "tin_tuc.csv"
output_file = "ketqua.csv"

# Từ điển các từ dừng (stop words)
stop_words = ["và", "là", "của"]  # Thêm các từ dừng vào danh sách

# Mở tệp CSV đầu vào
with open(input_file, "r", encoding="utf-8") as input_csv, \
     open(output_file, "w", newline="", encoding="utf-8") as output_csv:
    
    # Tạo reader và writer CSV
    reader = csv.reader(input_csv)
    writer = csv.writer(output_csv)
    header = next(reader)
    writer.writerow(header + ["entity","type","label","object"])
    
    # Duyệt qua từng hàng trong tệp CSV đầu vào
    for row in reader:
        text = row[0]  # cột đầu tiên chứa tiêu đề bài viết
        cleaned_text = re.sub(r"[,:%]", "", text) # xoá dấu
        # tiền xử lý văn bản
        cleaned_text = cleaned_text.lower()  # Chuyển đổi văn bản thành chữ thường
        tokens = word_tokenize(cleaned_text, format="text")  # Tách từ
        
        # loại bỏ các stop word
        tokens = [token for token in tokens.split() if token not in stop_words]
        
        # văn bản đã tiền xử lý
        preprocessed_text = " ".join(tokens)
        
        # Sử dụng hàm ner để nhận diện thực thể
        entities = ner(preprocessed_text)
        
        # Ghi thông tin từng thực thể vào tệp CSV đầu ra
        for entity in entities:
            writer.writerow(row + list(entity))  
```
